```python
#Importando a biblioteca Pandas
import pandas as pd
```

```python
#Leitura de arequivos
df1 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Datasets/Aracaju.xlsx")
df2 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Datasets/Fortaleza.xlsx")
df3 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Datasets/Natal.xlsx")
df4 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Datasets/Recife.xlsx")
df5 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Datasets/Salvador.xlsx")
```

```python
#Juntando todos os arquivos
df = pd.concat([df1, df2, df3, df4, df5])
```

```python
#Exibindo as 5 primeiras linhas
df.head()
```

```python
#Exibindo as 5 ltimas linhas
df.tail()
```

```python
#Verificando o tipo de cada coluna
df.dtypes
```

```python
#Alterando o tipo de dado da coluna LojaID
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes
```

```python
df.head()
```

```python
#Consultando linhas com valores faltantes
df.isnull().sum()
```

```python
#Substituindo os valores nulos pela mdia
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True)
```

```python
df["Vendas"].mean()
```

```python
df.isnull().sum()
```

```python
df.sample(15)
```

```python
#Substituindo valores nulos por Zero
df["Vendas"].fillna(0, inplace=True)
```

```python
#Apagando linhas com valores nulos
df.dropna(inplace=True)
```

```python
#Apagando linhas com valores nulos com base em apenas 1 coluna
df.dropna(subset=["Vendas"], inplace=True)
```

```python
#Removendo linhas que estejam com valores faltantes em todas as colunas
df.dropna(how="all", inplace=True)
```

## **Criando colunas novas**

```python
#Criando a coluna de receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
```

```python
df.head()
```

```python
#Retornando a maior receita
df["Receita"].max()
```

```python
#Retornando a menos receita
df["Receita"].min()
```

```python
#nlargest informa as maiores receitas, quantidade de linhas conforme parametro
df.nlargest(3, "Receita")
```

```python
#nsmallest informa as menores receitas, quantidade de linhas conforme parametro
df.nsmallest(3, "Receita")
```

```python
#Agrupamento por cidade - no caso fara a soma das vendas por cidades
df.groupby("Cidade")["Receita"].sum()
```

```python
#Ordenando o conjunto de dados
df.sort_values("Receita", ascending=False).head(10)
```

## **Trabalhando com datas**

```python
#Transformando a coluna data em tipo inteiro
df["Data"] = df["Data"].astype("int64")
```

```python
#Verificando o tupo de dado de cada coluna
df.dtypes
```

```python
#Transformando coluna de data em data
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
#Afrupamento por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#Criando uma nova coluna com o ano
df["Ano_venda"] = df["Data"].dt.year
```

```python
df.sample(5)
```

```python
#Extraindo o ms e o dia
df["mes_venda"], df["dia_venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```

```python
df.sample(5)
```

```python
#Retornando a data mais antiga
df["Data"].min()
```

```python
#Calculando a diferena de dias
df["diferenca_dias"] = df["Data"] - df["Data"].min()
```

```python
df.sample(5)
```

```python
#Criando a coluna de trimestre
df["semestre_venda"] = df["Data"].dt.quarter
```

```python
df.sample(5)
```

```python
#Criando a coluna de trimestre
df["trimestre_venda"] = df["Data"].dt.quarter
```

```python
df.sample(5)
```

```python
#Removendo coluna semestre_venda
df.drop(["semestre_venda"], axis=1, inplace=True)
```

```python
df.sample(5)
```

```python
#Filtrando as vendas de 2019 do ms de maro
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```
